```python
sns.countplot(data=df_drop, x='loan_type')
plt.title('Count of loan type')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df_drop, x='purpose_of_loan')
plt.title('Count of purpose of loan')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df_drop, x='employment_status')
plt.title('Count of employment status')
plt.xticks(rotation=45)
plt.show()
```

```python
# Weekly frequency
df_drop['application_date'] = pd.to_datetime(df_drop['application_date'])
df_drop.set_index('application_date', inplace=True)
weekly_loans = df_drop.resample('W').size().reset_index(name='loan_count')

plt.figure(figsize=(12, 6))
plt.plot(weekly_loans['application_date'], weekly_loans['loan_count'], marker='o')
plt.title('Weekly Loans Taken Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Loans')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
